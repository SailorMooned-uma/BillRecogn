```python
import numpy as np
import pandas as pd
from sklearn import datasets
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
ynames = iris.target_names[y]
```

```python
plt.figure(figsize=(15,5))
Z = hierarchy.linkage(X,method='ward')
dn = hierarchy.dendrogram(Z, color_threshold=7,labels=ynames)
plt.show()
```

```python
g = hierarchy.fcluster(Z,criterion='distance',t=7)
print(np.unique(g))
print(g)
```
