```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
wine = pd.read_csv('wine.csv')
df = pd.DataFrame(wine)
df
```

```python
df.describe()
```

```python
values = wine.describe()
values
```

```python
sns.countplot(df['quality'])
```

```python
cov_matrix = np.cov(df, rowvar=False)
plt.figure(figsize=(12, 12))
sns.heatmap(cov_matrix, annot=True)
#Covariance matrix
```

```python
corr_matrix = np.corrcoef(df, rowvar=False)
plt.figure(figsize=(12, 12))
sns.heatmap(corr_matrix, annot=True)
#Correlation matrix
```

```python
for i in df.columns:
    sns.boxplot(x=df[i])
    plt.title(i)
    plt.show()

    sns.boxplot(x='quality', y=i, data=df)
    plt.title(i + ' vs quality')
    plt.show()
```

```python
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['quality_scaled'] = scaler.fit_transform(df[['quality']])
fig = px.parallel_coordinates(df, color='quality_scaled', color_continuous_scale=px.colors.sequential.Reds)
fig.show()
```

```python
import pandas as pd
import plotly.express as px

dig = px.scatter(df, x="alcohol", y="pH", size="quality", color="quality", color_continuous_scale=px.colors.sequential.Blues)

dig.show()
```
