```python
iris = datasets.load_iris()
X = iris.data
y = iris.target_names
features = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
yy = np.resize(y, (len(X),1))
xx = np.concatenate((X,yy), axis=1)
df = pd.DataFrame(xx, columns=features)
df
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
tnames = iris.target_names
y = iris.target_names[iris.target]
features = np.concatenate((iris.feature_names, ['species']))
y = np.reshape(y, (len(X),1))
XX = np.concatenate((X,y), axis=1)
df = pd.DataFrame(XX, columns=features)
df[iris.feature_names] = df[iris.feature_names].astype(float)
df
```
